```python
# import the libraries
import warnings
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
# import the data and fix the column types
df_jpy_orig = pd.read_csv('Data/DEXJPUS_19710104_20180518.csv', header=0, index_col='DATE')

df_jpy_orig.index = pd.to_datetime(df_jpy_orig.index)
df_jpy_orig.DEXJPUS = pd.to_numeric(df_jpy_orig.DEXJPUS, errors='coerce')
df_jpy_orig = df_jpy_orig.fillna(method='ffill')
df_jpy_orig = df_jpy_orig.loc['2010':]

platinum_spread = 2 * 0.07 
gold_spread = 2 * 0.09
standard_spread = 2 * 0.15
```

```python
# on average, there's 21 trading days in a month
df_jpy = df_jpy_orig.copy()
df_jpy['21DMA'] = df_jpy['DEXJPUS'].rolling(window=21).mean()
df_jpy['21DMA_STD'] = df_jpy['DEXJPUS'].rolling(window=21).std()
df_jpy = df_jpy.dropna()

# 2*Standard Deviation is akin to creating a bollinger band
df_jpy['21DMA_UPPER_BAND'] = df_jpy['21DMA'] + 2*df_jpy['21DMA_STD']
df_jpy['21DMA_LOWER_BAND'] = df_jpy['21DMA'] - 2*df_jpy['21DMA_STD']

fig_jpy = df_jpy.plot(figsize=(16,5))
_ = fig_jpy.set_title('USDJPY Historical Price - Since 2010')
_ = fig_jpy.set_ylabel('Price')
_ = fig_jpy.set_ylim(ymin=60)
```

## Moving Average

```python
now = datetime.datetime.now()
```

```python
today_minus1Year = now.date() - datetime.timedelta(days=365)
df_jpy_past1Year = df_jpy.loc[today_minus1Year:]
fig_jpy_past1Year = df_jpy_past1Year.plot(figsize=(16,5), grid=True)
_ = fig_jpy_past1Year.set_title('USDJPY Historical Price - Past 1 Year')
_ = fig_jpy_past1Year.set_ylabel('Price')
_ = fig_jpy_past1Year.tick_params(labelright=True)
_ = fig_jpy_past1Year.set_ylim(ymin=df_jpy_past1Year['21DMA_LOWER_BAND'].min()-1, ymax=df_jpy_past1Year['21DMA_UPPER_BAND'].max()+1)
_ = fig_jpy_past1Year.fill_between(df_jpy_past1Year.index, df_jpy_past1Year['21DMA_LOWER_BAND'], df_jpy_past1Year['21DMA_UPPER_BAND'], alpha=0.1)
```

```python
df_jpy_past1Year['platinum_min_spread'] = platinum_spread
df_jpy_past1Year['gold_min_spread'] = gold_spread
df_jpy_past1Year['standard_min_spread'] = standard_spread
fig_jpy_past1Year_std = df_jpy_past1Year[['21DMA_STD','platinum_min_spread','gold_min_spread','standard_min_spread']].plot(figsize=(16,5), grid=True)
_ = fig_jpy_past1Year_std.set_ylabel('Standard Deviation')
_ = fig_jpy_past1Year_std.set_title('USDJPY Historical Price Standard Deviation - Past 1 Year')
```

## Relative Strength Index

```python
rsi_periods = 14; rsi_upper = 70; rsi_mid = 50; rsi_lower = 30
df_jpy = df_jpy_orig.copy()
df_jpy['DIFF'] = df_jpy.diff()
df_jpy = df_jpy.dropna()
df_jpy['GAIN'] = abs(np.where(df_jpy['DIFF']>=0, df_jpy['DIFF'], 0))
df_jpy['LOSS'] = abs(np.where(df_jpy['DIFF']<0, df_jpy['DIFF'],0))
df_jpy = df_jpy.rolling(window=rsi_periods).mean().dropna()[['GAIN','LOSS']]
df_jpy['RS'] = df_jpy['GAIN'] / df_jpy['LOSS']; df_jpy['RSI'] = 100 - (100 / (1 + df_jpy['RS']))
df_jpy['UPPER'] = rsi_upper; df_jpy['MID'] = rsi_mid; df_jpy['LOWER'] = rsi_lower
df_jpy_rsi_past1Year = df_jpy.loc[today_minus1Year:]
fig_jpy_rsi = df_jpy_rsi_past1Year[['RSI','UPPER','MID','LOWER']].plot(figsize=(16,5), title='Relative Strength Index - Past 1 Year', grid=True)
_ = fig_jpy_rsi.set_ylabel('RSI')
_ = fig_jpy_rsi.tick_params(labelright=True)
_ = fig_jpy_rsi.set_ylim(ymin=0, ymax=100)
#df_jpy_rsi_past1Year.head()
```
