```python
# import the libraries
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the data and fix the column types
df_jpy = pd.read_csv('Data/DEXJPUS_19710104_20180504.csv', header=0, index_col='DATE')
df_jpy.index = pd.to_datetime(df_jpy.index)
df_jpy.DEXJPUS = pd.to_numeric(df_jpy.DEXJPUS, errors='coerce')
df_jpy = df_jpy.fillna(method='ffill')
df_jpy = df_jpy.loc['2010':]
```

```python
df_jpy['30DMA'] = df_jpy['DEXJPUS'].rolling(window=30).mean()
df_jpy['30DMA_STD'] = df_jpy['DEXJPUS'].rolling(window=30).std()
df_jpy = df_jpy.dropna()
df_jpy['30DMA_UPPER_BAND'] = df_jpy['DEXJPUS'] + 2*df_jpy['30DMA_STD']
df_jpy['30DMA_LOWER_BAND'] = df_jpy['DEXJPUS'] - 2*df_jpy['30DMA_STD']

fig_jpy = df_jpy.plot(figsize=(16,5))
_ = fig_jpy.set_title('USDJPY Historical Price - Since 2010')
_ = fig_jpy.set_ylabel('Price')
_ = fig_jpy.set_ylim(ymin=60)
```

```python
now = datetime.datetime.now()
```

```python
today = now.date()
today_minus1Year = today - datetime.timedelta(days=365)

df_jpy_past1Year = df_jpy.loc[today_minus1Year:]
fig_jpy_past1Year = df_jpy_past1Year.plot(figsize=(16,5))
_ = fig_jpy_past1Year.set_title('USDJPY Historical Price - Past 1 Year')
_ = fig_jpy_past1Year.set_ylabel('Price')
_ = fig_jpy_past1Year.set_ylim(ymin=df_jpy_past1Year.DEXJPUS.min() - 10)
```
